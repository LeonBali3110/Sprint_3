## Crear nuevas columnas

```python
VelocidadMedia = (vuelos['Distance'] / vuelos['AirTime'])
```

```python
VelocidadMedia
```

```python
vuelos['VelocidadMedia'] = VelocidadMedia
```

```python
def etiquetar_retraso(Arr_Delay):
    if Arr_Delay == 0:
        return 'Sin retraso'
    elif Arr_Delay > 0:
        return 'Con retraso'
    else:
        return 'Llego más temprano'
```

```python
Retraso = vuelos['ArrDelay'].apply(etiquetar_retraso)
```

```python
Retraso
```

```python
vuelos['Retraso'] = Retraso
```

```python
vuelos
```
