Aquí creo la columna VelicidadMedia utilizando la formula 'distancia total / tiempo total' usando Distance y AirTime respectivamente. Ya que no estoy seguro de las unidades (especialmente de Distance) no las coloco.

```python
def etiquetar_retraso(Arr_Delay):
    if Arr_Delay <= 15:
        return 0
    else:
        return 1
```

```python
Retraso = vuelos['ArrDelay'].apply(etiquetar_retraso)
```

```python
Retraso
```

```python
vuelos['Retraso'] = Retraso
```

Realizo una función para determinar si los vuelos se retrasaron.
